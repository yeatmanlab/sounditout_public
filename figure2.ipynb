```python
# import necessary databases and libraries
import pandas as pd
import numpy as np
from scipy import stats
import seaborn as sns
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
plt.style.use('seaborn-whitegrid')
import numpy as np
from scipy.stats import linregress
from matplotlib import cm
#from colorspacious import cspace_converter
from collections import OrderedDict
cmaps = OrderedDict()
plt.rcParams['pdf.fonttype'] = 42
```

```python
# import data
word_data = pd.read_csv('data/study1/study1_word.csv')
pseudo_data = pd.read_csv('data/study1/study1_pseudo.csv')
acc_data = pd.read_csv('data/study1/study1_acc.csv')
rate_data = pd.read_csv('data/study1/study1_rate.csv')
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2,2, figsize = (12,12))

sns.violinplot(x="condition", y="acc",
               data = word_data, split=False, inner=None, palette=['darkgrey', 'royalblue'], ax=ax1)

for record in range(0, len(word_data.record_id.unique())):
    data = word_data[word_data.record_id == word_data.record_id.unique()[record]]
    data.groupby(['condition'])['acc'].mean().plot(kind='line', colormap='copper', ax=ax1)

ax1.set_title("Real word decoding")
ax1.set_xlabel("")
ax1.set_ylabel("Number correct")
ax1.set_xticklabels(["Uncued", "Cued"])


sns.violinplot(x="condition", y="acc",
               data = pseudo_data, split=True, inner=None, palette=['darkgrey', 'royalblue'], ax=ax2)

for record in range(0, len(pseudo_data.record_id.unique())):
    data = pseudo_data[pseudo_data.record_id == pseudo_data.record_id.unique()[record]]
    data.groupby(['condition'])['acc'].mean().plot(kind='line', colormap='copper',ax=ax2)

ax2.set_title("Pseudoword decoding")
ax2.set_xlabel("")
ax2.set_ylabel("Number correct")
ax2.set_xticklabels(["Uncued", "Cued"])    

sns.violinplot(x="condition", y="acc",
               data = acc_data, split=True, inner=None, palette=['darkgrey', 'royalblue'], ax=ax3)

for record in range(0, len(acc_data.record_id.unique())):
    data = acc_data[acc_data.record_id == acc_data.record_id.unique()[record]]
    data.groupby(['condition'])['acc'].mean().plot(kind='line', colormap='copper', ax=ax3)

ax3.set_title("Passage reading accuracy")
ax3.set_xlabel("Condition")
ax3.set_ylabel("Proportion correct")
ax3.set_xticklabels(["Uncued", "Cued"])

sns.violinplot(x="condition", y="rate",
               data = rate_data, split=True, inner=None, palette=['darkgrey', 'royalblue'],ax=ax4)

for record in range(0, len(rate_data.record_id.unique())):
    data = rate_data[rate_data.record_id == rate_data.record_id.unique()[record]]
    data.groupby(['condition'])['rate'].mean().plot(kind='line', colormap='copper', ax=ax4)

ax4.set_title("Passage reading rate")
ax4.set_xlabel("Condition")
ax4.set_ylabel("Correct words per second")
ax4.set_xticklabels(["Uncued", "Cued"])
```
